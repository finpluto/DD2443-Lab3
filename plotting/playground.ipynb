```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
local_df = pd.read_csv("../data/local_and_locked.log")
local_df["Distribution+OPS"] = local_df["Distribution"] + local_df["OPS"]
for set_name in ["Default", "Locked"]:
    sns.lineplot(data=local_df.loc[local_df["SetName"] == set_name], x="ThreadNumber", y="Time", hue="Distribution+OPS").set(title=set_name)
    plt.show()
```

```python
df = pd.read_csv("../data/total.txt", comment="#")
```

```python
draw_df2 = df
draw_df1 = post_df[post_df["SetName"] != "Default"]
g = sns.relplot(data=draw_df2, x="ThreadNumber", y="Time", hue="SetName", col="OPS", row="Distribution", kind="line")
```
